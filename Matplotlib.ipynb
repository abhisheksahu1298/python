```python
#Create labels for plot
#With Pyplot, We can use the xlabel () and ylabel() functions to set a label for the x-axis and y-axis.
# Example
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y, color="m", marker="o",linestyle="--",linewidth="2",markersize=10)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
#Create a Title for a plot
#With Pyplot, We can use the title() function to set a title for the plot.
# Example
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,"go:")

plt.title("Sports Watch Data",fontsize=20)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.legend(["average pulse"],loc=2)

plt.show()
```

```python
#Add Grid line to a plot
#With Pyplot, We can use the grid() function to  add grid lines to the plot.
# Example
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color="m")

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid(color='g',linestyle='--')

plt.show()
```

Pie Chart

```python
import matplotlib.pyplot as plt

classes=["Python","R","Java","DS","ML"]
class1_students=[30,20,45,20,55]

explode=(0,0,0.2,0,0)
colors=["c","b","r","y","m"]
textprops={"fontsize":15}
plt.pie(class1_students,labels= classes,explode=explode,colors=colors,autopct="%0.1f%%",shadow=True,textprops=textprops,frame= True,center=(2,3))
plt.show()
```

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5]
sleeping=[6,7,9,5,10]
eating=[2,3,1,5,4]
working=[8,7,9,7,10]
playing=[2,3,1,2,4]

plt.plot(color='m',label='sleeping',linewidth='4')
plt.plot(color='r',label='eating',linewidth='4')
plt.plot(color='b',label='working',linewidth='4')
plt.plot(color='y',label='playing',linewidth='4')

plt.stackplot(days,sleeping,eating,working,playing, colors=['m','r','b','y'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack plot')
plt.legend()
plt.show()
```
