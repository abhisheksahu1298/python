Pandas Series

```python
# Creating Pandas from a list
import pandas as pd
a=[1,2,3,4]
series1=pd.Series(a)
print(series1)
```

```python
# Labels
import pandas as pd
a=[1,2,3,4]
series1=pd.Series(a)
print(series1[1])
```

```python
#Creating a Pandas series from dictionary
import pandas as pd
calories= {"day1":200,"day2":250,"day3":350}
series2=pd.Series(calories)
print(series2)
```

Head Method

```python
# The head() method returns the headers and specified numbers of rows,starting from the top.

import pandas as pd
df=pd.read_csv('iplauction.csv')
print(df.head(10))
```

Tail Method

```python
#The tail() method returns the headers and specified numbers of rows,starting from the bottom.

import pandas as pd
df=pd.read_csv('iplauction.csv')
print(df.tail(10))
```

```python
#Print info about the data
#Example

import pandas as pd
df=pd.read_csv('iplauction.csv')
print(df.info())
```

Dropna method

```python
#The dropna()method returns a new Dataframe, and will not change the original.
#Return a new empty Dataframe with no empty cells

import pandas as pd
df=pd.read_csv('iplauction.csv')
new_df=df.dropna()

print(new_df.to_string())
```

```python
#If one want to change the original DataFrame, use the inplace=True.
import pandas as pd
df=pd.read_csv('iplauction.csv')
df.dropna(inplace=True)

print(df.to_string())
```

Fillna method

```python
#The fillna() method allows us to replace empty cells with a value.
#Example

import pandas as pd
df=pd.read_csv('iplauction.csv')
df.fillna(130, inplace=True)
```

Removing Duplicates

```python
#Return true for every row that is duplicate,else false.

print(df.duplicated())
```
